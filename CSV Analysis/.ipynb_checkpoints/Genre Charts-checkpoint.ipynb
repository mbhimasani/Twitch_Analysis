```python
import os
import pandas as pd
from glob import glob
import requests
import json
from pprint import pprint
import matplotlib.pyplot as plt

giant_key = "8f3214cefbfe33ef82a3137d0de164049a4aba85"
```

```python
filenames = glob('../Games CSV/ Games Data*.csv')
dataframe_list = [pd.read_csv(f) for f in filenames]
```

```python
streamers_files = glob('../Streamers CSV/ Stream Data*.csv')
data = [pd.read_csv(file) for file in streamers_files]
data_df = pd.concat(data)
streamer_df = data_df.groupby("Display Name").count()['Total Channel Views']
streamer_sorted = pd.DataFrame(streamer_df.sort_values(ascending=False))
streamer_sorted
```

```python
data_df
```

```python
all_data_df = pd.concat(dataframe_list)
all_data_df.head()
```

```python
import pybomb

my_key = giant_key
game_client = pybomb.GameClient(my_key)
giant_ids = list(all_data_df["Giantbomb ID"].unique())
giant_ids
```

```python
return_fields = ('id', 'name', 'genre')
genres = []
for game_id in giant_ids:
    print(game_id)
    if game_id == 0:
        genre = 'Other'
        genres.append(genre)
    else:
        response = game_client.fetch(game_id)
        try:
            genre = response.results['genres'][0]['name']
        except: KeyError
        print(genre)
        genres.append(genre)
print(genres)
```

```python
genres_df = pd.DataFrame({"GameID":giant_ids,
                        "Genre": genres})
genres_df
```

```python
giantbomb_ids = list(all_data_df["Giantbomb ID"])
df_genres = []
for giant_bomb_id in giantbomb_ids:
    df_genre = list(genres_df[genres_df['GameID']==giant_bomb_id]['Genre'])[0]
    df_genres.append(df_genre)
```

```python
all_data_df["Genre"] = df_genres
summary_genre = all_data_df.groupby('Genre').count()["Total Channels"]
summary_genre_df = pd.DataFrame(summary_genre)
genre_list = list(summary_genre_df.index.values)
all_data_df.to_csv("Game Data Including Genre.csv")
```

```python
genre_numbers = list(summary_genre_df['Total Channels'])
x_axis = list(range(0,len(genre_numbers)))
plt.barh(x_axis, genre_numbers, tick_label = genre_list)
plt.title("# of Times Each Genre Was in Top 25 Games by Views")
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
plt.show()
```

```python
import requests
import json
import pandas as pd
import numpy as np
import meghana_client_ID
from pprint import pprint
```

```python
url ="https://api.twitch.tv/kraken/games/top/?client_id="
client_ID = meghana_client_ID.meghana

query_url = url + client_ID
params = {"limit": "30"} #get top 30 games
```

```python
top_games = pd.DataFrame()
top_games["Date"]=""
top_games["Games"] = ""
headers = ["Rank", "Giantbomb id", "Viewer Count", "Number of Channels"]
for header in headers:
    top_games[header]=""
    top_games[[header]]=top_games[[header]].apply(pd.to_numeric, errors='ignore')

top_games.head()
```

```python
pprint(requests.get(query_url, params=params).json())
```

```python
list_25 = np.arange(0,25,1)
#for x in list_25:
topgame_response = requests.get(query_url, params=params).json()
for index, row in top_games.iterrows():
    print(query_url)
    top_games.loc[index, "Games"] = topgame_response['top'][index]['game']['name'] # get names of top 25 games
    top_games.loc[index, "Rank"] = index+1
    top_games.loc[index, "Giantbomb id"] = topgame_response['top'][index]['game']['giantbomb_id']
    top_games.loc[index, "Viewer Count"] = topgame_response['top'][index]['viewers']
    top_games.loc[index, "Number of Channels"] = topgame_response['top'][index]['channels']
    #top_games.loc[index, "Date"] = topgame_response['id'] # retrieve Date and add to df
#top_games.head()
```

```python
top_games[‘Date’] = pd.to_datetime(top_games[‘Date’],unit=‘s’)
```
